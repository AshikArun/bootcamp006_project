```python
import time
from selenium import webdriver
import numpy as np
import pandas as pd


driver = webdriver.Chrome('/Users/hanshuye/Downloads/chromedriver')  # Optional argument, if not specified will search path.
driver.get('https://www.tripadvisor.com/Hotels-g1499981-Krabi_Province-Hotels.html');
#time.sleep(5) # Let the user actually see something!

j=len(driver.find_elements_by_css_selector('.nav.next.ui_button.primary.taLnk'))

name_list=[]
id_list=[]
link_list=[]
rank_list=[]
star_list=[]

while (j>0):
    
    page_list=driver.find_elements_by_css_selector('.listing.easyClear.p13n_imperfect')
    for i in page_list:
        name=i.find_element_by_class_name('property_title').text
        ID=i.find_element_by_class_name('property_title').get_attribute("id")
        link=i.find_element_by_class_name('property_title').get_attribute("href")
        
        if (len(i.find_elements_by_class_name('slim_ranking'))>0):
            Rank=i.find_element_by_class_name('slim_ranking').text
        else:
            continue
        
        
        if (len(i.find_elements_by_class_name('sprite-ratings'))>0):
            Star=i.find_element_by_class_name('sprite-ratings').get_attribute("alt")
        else:
            continue
        
        name_list.append(name)
        id_list.append(ID)
        link_list.append(link)
        rank_list.append(Rank)
        star_list.append(Star)
        
    
    
    button_element=driver.find_element_by_css_selector('.nav.next.ui_button.primary.taLnk')
    button_element.click()
    time.sleep(5)
    j=len(driver.find_elements_by_css_selector('.nav.next.ui_button.primary.taLnk'))
    

page_list=driver.find_elements_by_css_selector('.listing.easyClear.p13n_imperfect')
for i in page_list:
        name=i.find_element_by_class_name('property_title').text
        ID=i.find_element_by_class_name('property_title').get_attribute("id")
        link=i.find_element_by_class_name('property_title').get_attribute("href")
        
        if (len(i.find_elements_by_class_name('slim_ranking'))>0):
            Rank=i.find_element_by_class_name('slim_ranking').text
        else:
            continue
        
        
        if (len(i.find_elements_by_class_name('sprite-ratings'))>0):
            Star=i.find_element_by_class_name('sprite-ratings').get_attribute("alt")
        else:
            continue
        
        name_list.append(name)
        id_list.append(ID)
        link_list.append(link)
        rank_list.append(Rank)
        star_list.append(Star)
         

basicinfo={'name':name_list,'id':id_list,'link':link_list,'rank':rank_list,'star':star_list}
basicinfo=pd.DataFrame(basicinfo)

basicinfo.to_pickle('/Users/hanshuye/Desktop/basic_info.pkl')

#i=1



#search_box = driver.find_element_by_class_name('property_title')
#print search_box.text
#time.sleep(5) # Let the user actually see something!
driver.quit()
```

```python
#id="ratingFilter" ul li label 2nd span text
#label for="taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_3" 2nd span text

#input id=taplc_prodp13n_hr_sur_review_filter_controls_0_filterSegment_Family
#label for=taplc_prodp13n_hr_sur_review_filter_controls_0_filterSegment_Family

#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSegment_Couples
#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSegment_Couples

#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSegment_Solo
#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSegment_Solo

#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSegment_Business
#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSegment_Business

#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSegment_Friends
#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSegment_Friends


#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSeasons_SPRING
#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSeasons_SPRING

#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSeasons_SUMMER
#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSeasons_SUMMER

#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSeasons_AUTUMN
#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSeasons_AUTUMN

#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSeasons_WINTER
#taplc_prodp13n_hr_sur_review_filter_controls_0_filterSeasons_WINTER



#div class="col language " ul 2nd li span input
#div class="col language " ul 2nd li label text
#div class="col language " ul 2nd li label span text

df = pd.read_pickle('/Users/hanshuye/Desktop/basic_info.pkl')
df=df[255:]
#finalframe=pd.DataFrame()
for index, row in df.iterrows():
    hotel_id=row['id']
    hotel_link=row['link']

    driver = webdriver.Chrome('/Users/hanshuye/Downloads/chromedriver')  # Optional argument, if not specified will search path.
    driver.get(hotel_link)

    purpose=['Family','Couples','Solo','Business','Friends']
    season=['SPRING','SUMMER','AUTUMN','WINTER']

    filter_head_seg='taplc_prodp13n_hr_sur_review_filter_controls_0_filterSegment_'
    filter_head_season='taplc_prodp13n_hr_sur_review_filter_controls_0_filterSeasons_'

    clean_list_seg=[]
    clean_list_season=[]

    for i in purpose:
        temp_element=driver.find_element_by_id(filter_head_seg+i)
        if (temp_element.is_selected()):
            clean_list_seg.append(i)

    for i in season:
        temp_element=driver.find_element_by_id(filter_head_season+i)
        if (temp_element.is_selected()):
            clean_list_season.append(i)

    for i in clean_list_seg:
        temp_element=driver.find_element_by_id(filter_head_seg+i)
        temp_element.click()
        time.sleep(2)

    for i in clean_list_season:
        temp_element=driver.find_element_by_id(filter_head_season+i)
        temp_element.click()
        time.sleep(2)

    #begin the iteration
    pre_purpose=""
    pre_season=""



    for i in purpose:
        if (pre_purpose!=""):
            driver.find_element_by_id(filter_head_seg+pre_purpose).click()
            time.sleep(2)

        driver.find_element_by_id(filter_head_seg+i).click()
        pre_purpose=i
        time.sleep(2)

        for j in season:
            if (pre_season!=""):
                driver.find_element_by_id(filter_head_season+pre_season).click()
                time.sleep(2)


            list_5=[]
            list_4=[]
            list_3=[]
            list_2=[]
            list_1=[]

            list_language=[]
            list_season=[]
            list_purpose=[]
            list_id=[]

            driver.find_element_by_id(filter_head_season+j).click()
            time.sleep(2)
            pre_season=j
            filter_lang=driver.find_element_by_css_selector('.col.language')
            filter_lang_list=filter_lang.find_elements_by_name('filterLang')
            if (len(filter_lang_list)>1):
                list_language.append(filter_lang_list[1].get_attribute("id"))
                if (not (filter_lang_list[1].is_selected())):
                    filter_lang_list[1].click()
                time.sleep(2)

                list_season.append(driver.find_element_by_id(filter_head_season+j).get_attribute("id"))
                list_purpose.append(driver.find_element_by_id(filter_head_seg+i).get_attribute("id"))

                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_5']/span)[2]")
                list_5.append(rate_value.text)
                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_4']/span)[2]")
                list_4.append(rate_value.text)
                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_3']/span)[2]")
                list_3.append(rate_value.text)
                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_2']/span)[2]")
                list_2.append(rate_value.text)
                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_1']/span)[2]")
                list_1.append(rate_value.text)

                list_id.append(hotel_id)
                #click on the second 

            filter_lang=driver.find_element_by_css_selector('.col.language')
            filter_lang_list=filter_lang.find_elements_by_name('filterLang')
            if (len(filter_lang_list)>2):
                list_language.append(filter_lang_list[2].get_attribute("id"))
                if (not (filter_lang_list[2].is_selected())):
                    filter_lang_list[2].click()
                time.sleep(2)

                list_season.append(driver.find_element_by_id(filter_head_season+j).get_attribute("id"))
                list_purpose.append(driver.find_element_by_id(filter_head_seg+i).get_attribute("id"))

                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_5']/span)[2]")
                list_5.append(rate_value.text)
                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_4']/span)[2]")
                list_4.append(rate_value.text)
                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_3']/span)[2]")
                list_3.append(rate_value.text)
                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_2']/span)[2]")
                list_2.append(rate_value.text)
                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_1']/span)[2]")
                list_1.append(rate_value.text)

                list_id.append(hotel_id)
                #click on the third 

            filter_lang=driver.find_element_by_css_selector('.col.language')
            filter_lang_list=filter_lang.find_elements_by_name('filterLang')
            if (len(filter_lang_list)>3):
                list_language.append(filter_lang_list[3].get_attribute("id"))
                if (not (filter_lang_list[3].is_selected())):
                    filter_lang_list[3].click()
                time.sleep(2)

                list_season.append(driver.find_element_by_id(filter_head_season+j).get_attribute("id"))
                list_purpose.append(driver.find_element_by_id(filter_head_seg+i).get_attribute("id"))

                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_5']/span)[2]")
                list_5.append(rate_value.text)
                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_4']/span)[2]")
                list_4.append(rate_value.text)
                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_3']/span)[2]")
                list_3.append(rate_value.text)
                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_2']/span)[2]")
                list_2.append(rate_value.text)
                rate_value=driver.find_element_by_xpath("(//label[@for='taplc_prodp13n_hr_sur_review_filter_controls_0_filterRating_1']/span)[2]")
                list_1.append(rate_value.text)

                list_id.append(hotel_id)
                #click on the fourth

            temp_frame=pd.DataFrame({'id':list_id,'purpose':list_purpose,'season':list_season,'language':list_language,'1':list_1,'2':list_2,'3':list_3,'4':list_4,'5':list_5})
            finalframe=pd.concat([finalframe,temp_frame])

    driver.quit()


```
