```python
import time
from selenium import webdriver
import numpy as np
import pandas as pd
rating_frame = pd.read_pickle('/Users/hanshuye/Desktop/rating_frame.pkl')
```

```python
rating_frame.head(n=10)
```

```python
rating_frame=rating_frame.reset_index()
```

```python
rating_frame=rating_frame[(rating_frame['1']+rating_frame['2']+rating_frame['3']+rating_frame['4']+rating_frame['5'])!="00000"]
```

```python
len(rating_frame)
```

```python
int_list=['1','2','3','4','5']
for i in int_list:
    rating_frame[i]=map(int,rating_frame[i])
```

```python
cut_list=['language','purpose','season']
for i in cut_list:
    rating_frame[i]=map(lambda x:x[x.rfind('_')+1:],rating_frame[i])
```

```python
rating_frame.head(n=10)
```

```python
rating_frame['sum']=rating_frame['1']+rating_frame['2']+rating_frame['3']+rating_frame['4']+rating_frame['5']
rating_frame.head(n=10)
```
